```python
import numpy as np
```

```python
X=np.array([1,3,5,7]) #experience
y=np.array([3,6,9,12]) #salary (LPA)
```

```python
sum_X=np.sum(X)
sum_y=np.sum(y)
sum_X2=np.sum(X**2)
sum_Xy=np.sum(X*y)
```

```python
n=len(X)
```

```python
m=(n*sum_Xy-sum_X*sum_y)/(n*sum_X2-sum_X**2)
```

```python
c=(sum_y-m*sum_X)/n
```

```python
print("Slope:",m)
print("Intercept:",c)
```

```python
experience=10
predicted_salary=m*experience+c
print("Predicted salary:",predicted_salary)
```
