```python
import numpy as np
from collections import Counter

```

```python
X_train = np.array([
    [2, 120, 70, 25.3, 31],
    [0, 95, 60, 22.0, 25],
    [3, 130, 80, 28.1, 35],
    [1, 85, 65, 20.0, 22],
    [5, 160, 90, 32.0, 45],
    [4, 140, 85, 30.5, 40],
    [0, 100, 70, 23.0, 28],
    [6, 170, 95, 35.0, 50]
])

# Labels: 1 = Diabetes, 0 = No Diabetes
y_train = np.array([1, 0, 1, 0, 1, 1, 0, 1])
```

```python
X_test=np.array([6,81,44,46.4,82])
K=3
```

```python
distances = []
```

```python
for i in range(len(X_train)):
 distance = np.sqrt(np.sum(X_train[i]-X_test)**2)
 distances.append((distance,y_train[i]))
```

```python
# sort
distances.sort(key=lambda X:X[0])
```

```python
K_nearest_label =(label for distance,label in distances[:K])
```

```python
# prediction
prediction = Counter(K_nearest_label).most_common(1)[0][0]
```

```python
if prediction==1:
  print("Patient has Diabetes")
else:
  print("Patient does not have Diabetes")


```
